**Why do electronics, books, and sports have such good average ratings and also Who do Clothing Toys have such low average ratings?**

**Electronics**

Facts over feelings: People rate these based on clear specs like battery life or speed, which are easy for brands to get right.

Smart shoppers: Most people (68%+) do heavy research first, so they almost always end up with a product that fits their needs.

Big brand trust: Household names like Apple or Sony have mastered quality control and offer solid warranties that keep people happy.
The "Wow" factor: Getting a new, faster gadget creates a strong positive first impression that leads to immediate 5-star ratings.

**Books**

Self-filtering: You usually only buy a book you already think you'll like, which naturally keeps ratings high.

Superfans: Loyal readers often drop 5-star reviews for their favorite authors the second a book comes out.

Better matchmaking: Apps like Goodreads help people find exactly what they enjoy, so there are fewer "wrong" purchases.

Kindness bias: In the book world, a 3-star is often seen as "okay," so most people default to 4 or 5 stars if they finished the book.

**Sports & Fitness**

Brand love: People feel a deep connection to brands like Nike or Adidas because of the athletes or teams they represent.

Feel-good vibes: Since these products are tied to health and self-improvement, people associate them with a positive lifestyle.

Niche perfection: Brands now make gear for very specific hobbies (like hot yoga or trail running), so the "fit" for the user is almost always perfect.

Community spirit: Wearing team gear makes people feel like they belong to a group, which boosts their overall satisfaction with the purchase.

**Clothing**

The sizing gamble: Sizes vary wildly between brands, leading to huge frustration when things don't fit right.

Touch vs. Tech: You can't feel the fabric online, so people often end up disappointed by "cheap" or "scratchy" materials.

Color tricks: Photos often look better than reality because screen settings change how colors appear in person.

Bad timing: People usually buy clothes for specific events; if a dress or suit arrives late, it's an automatic 1-star review.

**Toys**

Built to break: Kids are rough on toys, so anything that feels like "cheap plastic" or snaps easily gets slammed in reviews.

Safety first: Parents are (rightfully) obsessed with safety; any sharp edges or weird smells lead to instant negative feedback.

Hype vs. Reality: If a toy doesn't do the "cool tricks" shown in the commercials, parents feel like they've been ripped off.

Missing pieces: With so many tiny parts, it’s common for sets to arrive incomplete, which ruins the fun for the kid and the paren

```python
import matplotlib.pyplot as plt
import math

# 1. Prepare the data
month_cols = [col for col in df.columns if col.startswith('sales_month_')]
categories = df['category'].unique()
months = [f"M{i+1}" for i in range(len(month_cols))]

# 2. Setup the grid (e.g., 2 columns, rows based on number of categories)
cols = 2
rows = math.ceil(len(categories) / cols)
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 5))
axes = axes.flatten() # Flatten to easily loop through them

for i, cat in enumerate(categories):
    cat_df = df[df['category'] == cat]
    
    # Calculate Sales and Revenue for this category
    cat_sales = cat_df[month_cols].sum()
    cat_rev = cat_df[month_cols].multiply(cat_df['price ($)'], axis=0).sum()
    
    # Plot Sales (Left Axis)
    ax1 = axes[i]
    ax1.plot(months, cat_sales, color='tab:blue', marker='o', label='Sales (Units)')
    ax1.set_ylabel('Sales (Units)', color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    ax1.set_title(f'Category: {cat}', fontweight='bold')
    
    # Plot Revenue (Right Axis)
    ax2 = ax1.twinx()
    ax2.plot(months, cat_rev, color='tab:red', marker='s', linestyle='--', label='Revenue ($)')
    ax2.set_ylabel('Revenue ($)', color='tab:red')
    ax2.tick_params(axis='y', labelcolor='tab:red')

# 3. Clean up empty subplots and layout
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

```

Seasonal Sales Analytics

```python
winter_cols = ['sales_month_12', 'sales_month_1', 'sales_month_2']
spring_cols = ['sales_month_3', 'sales_month_4', 'sales_month_5']
summer_cols = ['sales_month_6', 'sales_month_7', 'sales_month_8']
fall_cols   = ['sales_month_9', 'sales_month_10', 'sales_month_11']
```

```python
# Create seasonal totals
df['Winter'] = df[winter_cols].sum(axis=1)
df['Spring'] = df[spring_cols].sum(axis=1)
df['Summer'] = df[summer_cols].sum(axis=1)
df['Fall']   = df[fall_cols].sum(axis=1)

# Group by category
usa_seasonal_sales = df.groupby('category')[['Winter', 'Spring', 'Summer', 'Fall']].sum()
print(usa_seasonal_sales)
```

```python
seasonal_summary = df.groupby('category')[['Winter', 'Spring', 'Summer', 'Fall']].sum()

# 3. Create the Bar Graph
ax = seasonal_summary.plot(kind='bar', figsize=(12, 6), width=0.8)

# 4. Customise labels and title
plt.title('Seasonal Sales by Category (USA)', fontsize=14)
plt.ylabel('Total Units Sold')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.legend(title='Seasons')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python

# 1. Calculate Seasonal Revenue (Sales * Price)
df['Winter Revenue'] = df[['sales_month_12', 'sales_month_1', 'sales_month_2']].sum(axis=1) * df['price ($)']
df['Spring Revenue'] = df[['sales_month_3', 'sales_month_4', 'sales_month_5']].sum(axis=1) * df['price ($)']
df['Summer Revenue'] = df[['sales_month_6', 'sales_month_7', 'sales_month_8']].sum(axis=1) * df['price ($)']
df['Fall Revenue']   = df[['sales_month_9', 'sales_month_10', 'sales_month_11']].sum(axis=1) * df['price ($)']

# 2. Group by Category
seasonal_rev_summary = df.groupby('category')[['Winter Revenue', 'Spring Revenue', 'Summer Revenue', 'Fall Revenue']].sum()

# 3. Create the Bar Graph
ax = seasonal_rev_summary.plot(kind='bar', figsize=(12, 6), color=['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728'])

# 4. Formatting for clarity
plt.title('Seasonal Revenue by Category ($)', fontsize=14, fontweight='bold')
plt.ylabel('Total Revenue ($)')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.legend(title='Seasons', labels=['Winter', 'Spring', 'Summer', 'Fall'])
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Optional: Format the Y-axis to show "Millions" or "Thousands" if numbers are huge
from matplotlib.ticker import StrMethodFormatter
ax.yaxis.set_major_formatter(StrMethodFormatter('${x:,.0f}'))

plt.tight_layout()
plt.show()
```
